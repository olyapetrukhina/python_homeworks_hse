```python
import nltk
nltk.download('wordnet')

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

from nltk import download
download('punkt')

download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
def text_clean(text):
  text = text.lower()
  text = word_tokenize(text)
  wordnet_lemmatizer = WordNetLemmatizer()
  text = [word for word in text if word[0].isalpha()]
  text = [word for word in text if word not in stop_words]
  text_clean = [wordnet_lemmatizer.lemmatize(word) for word in text]

  return text_clean

text = '''Soft kitty, warm kitty,
Little ball of fur
Happy kitty, sleepy kitty,
Purr, purr, purr'''

print(text_clean(text))
```

```python
full_text = text_clean(text)
print(full_text[:20])

from collections import Counter
Counter(full_text).most_common(3)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wordcloud = WordCloud().generate(', '.join(full_text))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
freq_bigramms = Counter(nltk.bigrams(full_text))
freq_bigramms.most_common(5)
```
